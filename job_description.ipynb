```python
import pandas as pd
import os
import string
import re
import nltk

```

```python
job_dataset = pd.read_csv(r"/home/dell/Desktop/Skill_gap/job_descriptions.csv")
# job_dataset.head()
```

```python
missing_values = job_dataset.isnull().sum()
# display("Missing Values:\n", missing_values)
```

```python
job_dataset = job_dataset.dropna()
# display(job_dataset)
```

```python
# Example: Convert 'Job Posting Date' to datetime format
job_dataset['Job Posting Date'] = pd.to_datetime(job_dataset['Job Posting Date'])
```

```python
# removed latitude and longitude columns,since we have country
job_dataset = job_dataset.drop(['latitude', 'longitude', 'Contact'], axis=1)
# display(job_dataset)
```

```python
# covert all data to lower case
job_dataset = job_dataset.applymap(lambda x: x.lower() if isinstance(x, str) else x)
# display(job_dataset)
```

```python
def remove_punctuation_from_cell(cell):
    translator = str.maketrans("", "", string.punctuation)
    return str(cell).translate(translator)

for column in job_dataset.columns:
    if  job_dataset[column].dtype == 'object': # Check if the column contains text
        job_dataset[column] = job_dataset[column].apply(remove_punctuation_from_cell)
        
#         display(job_dataset)
```

```python
  # adding a standardized format with a hyphen in Salary Range Column
def clean_salary_range(salary_range):
    min_salary = salary_range[:2]
    max_salary = salary_range[3:]
    
    #adding a standardized format with a hyphen
    standardized_salary_range = f"{min_salary}-{max_salary}"
    
    return standardized_salary_range

job_dataset['Salary Range'] = job_dataset['Salary Range'].apply(clean_salary_range)
# display(job_dataset)
```

```python
# putting spaces b2n words in Compony profile column
def add_spaces(text):
    clean_text = re.sub(r'(sector|industry|other_keyword)', r'\1 ', text)
    
    return clean_text

job_dataset['Company Profile'] = job_dataset['Company Profile'].apply(add_spaces)

def add_space_before_industry(text):
    return text.replace('industry', ' industry')

job_dataset['Company Profile'] = job_dataset['Company Profile'].apply(add_space_before_industry)
# display(job_dataset)
```

```python
# Have created a version of my dataset, of the cleaned data
job_dataset.to_csv('cleaned_data.csv', index=False)
```

```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Assuming 'text_column' is the name of the column containing the text data
text_column = 'skills'  # Replace 'your_column_name' with the actual column name

# Download NLTK stop words
stop_words = set(stopwords.words('english'))

# Function to remove stop words from a text
def remove_stopwords(text):
    words = word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

# Apply the function to the specified column
tokenized_data[text_column] = tokenized_data[text_column].apply(remove_stopwords)

# Display the updated DataFrame
tokenized_data.head()
```

```python
# Assuming 'Skills' is the column containing skills information
skills_column = clean_data['skills']

# Initialize the TF-IDF vectorizer
tfidf_vectorizer_skills = TfidfVectorizer()

# Fit and transform the text data
X_tfidf_skills = tfidf_vectorizer_skills.fit_transform(skills_column)

# X_tfidf_skills is now a sparse matrix containing TF-IDF vectors for skills
```

```python
# Assuming 'Job Title' is the column containing job titles
job_title_column = clean_data['Job Title']

# Initialize the TF-IDF vectorizer
tfidf_vectorizer_job_title = TfidfVectorizer()

# Fit and transform the text data
X_tfidf_job_title = tfidf_vectorizer_job_title.fit_transform(job_title_column)

# X_tfidf_job_title is now a sparse matrix containing TF-IDF vectors for job titles
```

```python
# Assuming 'Qualifications' is the column containing qualifications information
qualifications_column = clean_data['Qualifications']

# Initialize the TF-IDF vectorizer
tfidf_vectorizer_qualifications = TfidfVectorizer()

# Fit and transform the text data
X_tfidf_qualifications = tfidf_vectorizer_qualifications.fit_transform(qualifications_column)

# X_tfidf_qualifications is now a sparse matrix containing TF-IDF vectors for qualifications
```

```python
import pandas as pd

# Get feature names from the TF-IDF vectorizer
feature_names_responsibilities = tfidf_vectorizer.get_feature_names_out()
feature_names_skills = tfidf_vectorizer_skills.get_feature_names_out()
feature_names_job_title = tfidf_vectorizer_job_title.get_feature_names_out()
feature_names_qualifications = tfidf_vectorizer_qualifications.get_feature_names_out()

# Convert TF-IDF matrices to DataFrames
df_tfidf_responsibilities = pd.DataFrame(X_tfidf.toarray(), columns=feature_names_responsibilities)
df_tfidf_skills = pd.DataFrame(X_tfidf_skills_dense, columns=feature_names_skills)
df_tfidf_job_title = pd.DataFrame(X_tfidf_job_title_dense, columns=feature_names_job_title)
df_tfidf_qualifications = pd.DataFrame(X_tfidf_qualifications_dense, columns=feature_names_qualifications)

# Display DataFrames
print("TF-IDF DataFrame for Responsibilities:")
print(df_tfidf_responsibilities)

print("\nTF-IDF DataFrame for Skills:")
print(df_tfidf_skills)

print("\nTF-IDF DataFrame for Job Title:")
print(df_tfidf_job_title)

print("\nTF-IDF DataFrame for Qualifications:")
print(df_tfidf_qualifications)
```
